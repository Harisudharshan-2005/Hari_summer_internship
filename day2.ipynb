```python
def run_query(query, db_name='college.db'):
  conn = sqlite3.connect(db_name)
  data = pd.read_sql(query, conn)
  conn.close()
  print(f"=== Data from {db_name} Loaded Successfully ===")
  print("")
  return data
```

```python
query = "SELECT student_id , name FROM college LIMIT 5"
print(run_query(query))
```

```python
query = "SELECT name , department , math_score FROM college WHERE math_score>=90 ORDER BY math_score DESC LIMIT 5"
print(run_query(query))
```

```python
query = "SELECT name , math_score , science_score , programming_score , attendance_percentage FROM college WHERE department = 'Computer Science' ORDER BY programming_score DESC"
print(run_query(query))
```

```python
query = "SELECT name ,department , attendance_percentage FROM college WHERE department != 'Civil' AND attendance_percentage > 90 ORDER BY attendance_percentage DESC"
print(run_query(query))
```

```python
query = "SELECT name ,department , attendance_percentage FROM college WHERE department != 'Mechanical' AND attendance_percentage > 93 ORDER BY attendance_percentage DESC"
print(run_query(query))
```
