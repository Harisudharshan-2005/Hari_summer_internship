```python
def run_query(query, db_name='college.db'):
  conn = sqlite3.connect(db_name)
  data = pd.read_sql(query, conn)
  conn.close()
  print(f"=== Data from {db_name} Loaded Successfully ===")
  print("")
  return data
```

```python
query = "SELECT student_id , name FROM college LIMIT 5"
print(run_query(query))
```

```python
conn = sql.connect('college.db')
cursor = conn.cursor()
chart_sql = """
SELECT department , ROUND(AVG(math_score)) AS  avg_math
FROM college
GROUP BY department
ORDER BY avg_math DESC
"""
data = pd.read_sql(chart_sql,conn)
conn.close()
print("=== DataBase Connected Successfully ===")
print("")
print(data)
conn.close
```

```python
fig, ax = plt.subplots(figsize=(10,8 ))
plt.xlabel("DEPARTMENT")
plt.ylabel("AVERAGE MATH SCORE")
plt.title("DEPARTMENT OF AVERAGE MATH SCORE")
ax.bar(data['department'],data['avg_math'], color =['red','yellow','black','blue'])
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
conn = sql.connect('college.db')
cursor = conn.cursor()
chart_sql = """
SELECT name , (math_score + programming_score + math_score + science_score + english_score) AS total_score
FROM college
ORDER BY total_score
LIMIT 8;
"""
data = pd.read_sql(chart_sql,conn)
print("=== DataBase Connected Successfully ===")
print("")
print(data)
conn.close()
```

```python
fig , ax = plt.subplots(figsize=(10,8))
plt.title("TOP 8 STUDENTS WITH TOP SCORES IN ALL THE DEPARTMENT")
plt.xlabel("STUDENTS")
plt.ylabel("TOTAL SCORE")
plt.barh(data['name'],data['total_score'], color = ['blue','black','orange','yellow','red','blue','grey','gold'])
plt.show()
```

```python
import sqlite3 as sql
import pandas as pd

conn = sql.connect('college.db')
chart_sql = """
SELECT department, COUNT(student_id) AS student_count
FROM college
GROUP BY department
"""
department_data = pd.read_sql(chart_sql, conn)
conn.close()

# Count students department-wise
dept_count = department_data.set_index('department')['student_count']

# Create pie chart
fig, ax = plt.subplots(figsize=(8,8))

plt.pie(
    dept_count.values,
    labels=dept_count.index,
    autopct='%1.1f%%',
    colors=['red', 'blue', 'purple', 'orange']
)

plt.title("STUDENT DISTRIBUTION BY DEPARTMENT")

plt.show()
```

```python
import matplotlib.pyplot as plt
conn = sql.connect('college.db')
cursor = conn.cursor()
chart_sql = """
SELECT gender , ROUND(AVG(programming_score)) AS avg_programming_score
FROM college
WHERE gender = "Female"
GROUP BY gender
ORDER BY avg_programming_score DESC
"""
data = pd.read_sql(chart_sql,conn)
print("=== DataBase Connected Successfully ===")
print("")
print(data)
conn.close()
```

```python
fig , ax = plt.subplots(figsize=(2,7))
plt.title("AVERAGE PROGRAMMING SCORE OF FEMALE")
plt.xlabel("GENDER")
plt.ylabel("AVERAGE PROGRAMMING SCORE")
plt.bar(data['gender'],data['avg_programming_score'], color = ['gold'])
plt.show()
```

```python
import matplotlib.pyplot as plt
conn = sql.connect('college.db')
cursor = conn.cursor()
chart_sql = """
SELECT department , AVG(attendance_percentage) AS avg_attendance
FROM college
GROUP BY department
HAVING avg_attendents>85
ORDER BY attendance_percentage DESC>85
"""
data = pd.read_sql(chart_sql,conn)
print("=== DataBase Connected Successfully ===")
print("")
print(data)
conn.close()
```
